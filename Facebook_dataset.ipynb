```python
class Coder():
    def __init__(self):
        self.code = 0
        self.dct = {}
    def __call__(self,word):
        if word in self.dct.keys():
            return self.dct[word]
        else:
            self.code += 1
            self.dct[word] = self.code
            return self.code
conv = {1: Coder()}
dataset = np.genfromtxt(fname,delimiter=';',names=True,converters=conv,filling_values=0)
pandas_df = pd.DataFrame(dataset)
```

#### Для опредедения наиболее популярного поста я выберу небольшое число показателей, домустим 4, и определю метрику как взвешенную сумму.
Самое главное - выбор правильных показателей.

1 предположение: если пост популярен, то большая часть увидевших его пользователей совершит какое-то действие (лайк, коммент, репост). Первый признак = Total_Interactions/Lifetime_Post_Consumers. Это отношение числа реальных действий к числу пользователей, которые хоть раз кликнули на пост
2 предположение:  популярный пост должен задеть большое количество новых людей, не просматривающих страницу регулярно. Это будет говорить о том, что пост вышел за пределы обычной аудитории. Объем регулярной аудитории оценивается как число людей, поставивших лайк на странице. Напрямую это можно было бы определить на много проще, если бы был известен средний охват пользователей за предыдущее время постами страницы. 

Косвенно это можно оценить как Lifetime_Post_Total_Reach - Lifetime_Post_reach_by_people_who_like_your_Page и

Lifetime_Post_Total_Impressions - Lifetime_Post_Impressions_by_people_who_have_liked_your_Page

Это оценка того, на сколько сильно пост вышел за пределы обычной аудитории
Последним учитываемым признаком будет охват аудитории Lifetime_Post_Total_Reach

Все показатели будут нормированы

```python
effective_reach = dataset['Total_Interactions']/dataset['Lifetime_Post_Consumers']

total_reach = dataset['Lifetime_Post_Total_Reach']
regular_reach = dataset['Lifetime_Post_reach_by_people_who_like_your_Page']
overreach = (total_reach - regular_reach)/total_reach

total_impress = dataset['Lifetime_Post_Total_Impressions']
regular_impress = dataset['Lifetime_Post_Impressions_by_people_who_have_liked_your_Page']
over_impress = (total_impress - regular_impress)/total_impress
reach = dataset['Lifetime_Post_Total_Reach']/np.max(dataset['Lifetime_Post_Total_Reach'])

```

Стакаем массивы с показателями и считаем метрику как взвешенную сумму через скалярное произведение

```python
ds_features = np.stack((np.arange(dataset.shape[0],dtype=np.int),reach,effective_reach,overreach,over_impress),1)
coefficients = np.array([0,1,4,3,2])
score = np.dot(ds_features,coefficients)
score_id = np.stack((np.arange(dataset.shape[0],dtype=np.int),score),1)
```
